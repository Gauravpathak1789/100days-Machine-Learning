ORDINAL ENCODING---

```python
from sklearn.preprocessing import OrdinalEncoder
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(df1.iloc[:,:2],df1.iloc[:,2:],test_size=0.2)
```
